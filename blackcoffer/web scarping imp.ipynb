```python
def remove_pattern(text,pattern):
    x=re.findall(pattern,str(text))
    y=str(text)
    for i in x:
        y=re.sub(i,'',str(y))
    return y
```

## SYLLABLE PER WORD


```python
def count_syllables(word):
    vowels = "aeiouy"
    syllables = 0
    last_was_vowel = False
    for char in word:
        if char.lower() in vowels:
            if not last_was_vowel:
                syllables += 1
            last_was_vowel = True
        else:
            last_was_vowel = False
            if char.lower() == "e":
                if syllables == 0:
                    syllables += 1
                break
    if syllables == 0:
        syllables = 1
    return syllables
```

```python
syllable_count = []
for i in range(0, len(df)):
    x = nltk.word_tokenize(df['Tokenized'][i])
    y = [count_syllables(word) for word in x]
    syllable_count.append(y)
```
